```python
#accept employee basic salary and find gross salary

basic=int(input("enter your basic salary to find gross salary"))
if basic<=10000:
   print()  #HRA is 70% on basic , DA is 56% on basic
   hra=70*basic
   da=56*basic
elif basic<=20000:
    print() #HRA is 73% on basic , DA is 60% on basic
    hra=73*basic
    da=60*basic
else:
    print() #HRA is 80% on basic , DA is 56% on basic
    hra=80*basic
    da=56*basic

gs=basic+hra+da
print("gross salary:",gs)
```
