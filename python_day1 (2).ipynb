```python
#1.accept project marks, internal marks, external marks from students
#2.passmarks is 50 in each subject
#3.calculate total score if student got more than or equal to 50 in each subject.

#total_score=70% on project + 20% on external
#                 +
              #10% on internal
#print A grade if score is 90+
#B grade if score is b/w 70-90
#C grade if score is b/w 50-70

project=int(input("enter project marks:"))
internal=int(input("enter internal marks:"))
external=int(input("enter external marks:"))
if project>=50 and internal>=50 and external>=50:
  total=(project/100)*70+(external/100)*20+(internal/100)*10
  print("total score:",total)
  if total>=90:
    print("A grade")
  elif total>=70:
    print("B grade")
  else:
    print("C grade")
else:
  print("fail")
```

```python
#accept start and stop value and print b/w values
#start: 10
#stop: 100
#o/p: 10 11 12...99

#start:  100
#stop: 10
#o/p: 100 99 98 ...10

#start:10
#stop:10
#o/p: your start and stop values are same

start=int(input("enter start value:"))
stop=int(input("enter stop value:"))
if start<stop:
   for i in range(start,stop):
       print(i,end=" ")
elif start>stop:
  for i in range(start,stop,-1):
    print(i,end=" ")
else:
  print("your start and stop values are same")
```

```python
# Write a program to print factors of a given number
# 10  --> 1 2 5
# 50  --> 1 2 5 10 25
num=int(input("Enter your number to print factors :"))
for i in range(1, (num//2)+1):
    if num%i==0:
        print(i,end=" ")

```

```python
# Write a program to print count of factors of a given number
count=0
num=int(input("Enter your number to print count of factors :"))
for i in range(1, (num//2)+1):
    if num%i==0:
        count=count+1

print("number of factors: ",count)
```

```python
# Write a program to check given number is prime or not
# 2 3 5 7 11 13 17 19 ....
count=0
num=int(input("Enter your number to check it is prime or not :"))
for i in range(1, (num//2)+1):
    if num%i==0:
        count=count+1

if count==1:
    print(num," is a prime number")
else:
    print(num," is not a prime number")
```

```python
# Write a program to check given number is perfect number or not
# 6      :  1+2+3  ==6
# 28     : 1+2+4+7+14===28
# 10     : 1+2+5  ===> 8

sum=0
num=int(input("Enter your number to check it is perfect number or not :"))
for i in range(1, (num//2)+1):
    if num%i==0:
        sum=sum+i

if sum==num:
    print(num," is perfect number")
else:
    print(num," is not a perfect number")
```
